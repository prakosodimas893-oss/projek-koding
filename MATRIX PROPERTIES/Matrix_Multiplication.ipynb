# Matrix Multiplication

```python
import numpy as np
```

```python
# buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy
A = np.array([[2, 5, 1], [5, 6, 1],])
A
```

```python
b = np.array([[6],
              [1],
              [4]])
b
```

```python
import torch
```

```python
# buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
A_pt = torch.tensor([[1, 5, 6], [1, 4, 2]])
A_pt
```

```python
b_pt = torch.tensor([[2],
                     [4],
                     [1]])
b_pt
```

```python
import tensorflow as tf
```

```python
# buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
A_tf = tf.Variable([[1, 2, 3], [4, 5, 6]])
A_tf
```

```python
b_tf = tf.Variable([[7],
                    [8],
                    [9]])
b_tf
```

```python
# tampilkan ukuran matrix A dan vector b
X = A.shape
X
```

```python
X = b.shape
X
```

```python
# kalikan matrix A dengan vector b
np.dot (A, b)
```

```python
# kalikan matrix A_pt dengan vector b_pt
torch.matmul(A_pt, b_pt)
```

```python
# kalikan matrix A_tf dengan vector b_tf
tf.linalg.matvec(A_tf, tf.squeeze(b_tf))
```

```python
# buat matrix B dengan ukuran 3x3 menggunakan numpy
B = np.array([[2, 5, 1], [5, 6, 1], [4, 0, 9]])
B
```

```python
# buat matrix B_pt dengan ukuran 3x3 menggunakan pytorch
B_pt = torch.tensor([[2, 5, 1], [5, 6, 1], [4, 0, 9]])
B_pt
```

```python
# buat matrix B_tf dengan ukuran 3x3 menggunakan tensorflow
B_tf = tf.Variable([[2, 5, 1], [5, 6, 1], [4, 0, 9]])
B_tf
```

```python
# kalikan matrix A dengan matrix B
np.dot(A, B)
```

```python
# kalikan matrix A_pt dengan matrix B_pt
torch.matmul(A_pt, B_pt)
```

```python
# kalikan matrix A_tf dengan matrix B_tf
tf.linalg.matmul(A_tf, B_tf)
```

```python
# buktikan apakah perkalian matrix bersifat komutatif? (AB=BA)?
# np.dot(B, A)
```

```python
B_pt = torch.from_numpy(B)
B_pt
```

```python
B_pt = torch.tensor([[2, 5, 1], [5, 6, 1], [4, 0, 9]]).T
B_pt
```

```python
torch.matmul(A_pt, B_pt)
```

```python
B_tf = tf.convert_to_tensor(B, dtype=tf.int32)
B_tf
```

```python
tf.matmul(A_tf, B_tf)
```
