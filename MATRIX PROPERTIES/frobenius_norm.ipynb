#Frobenius Norm

```python
# menggunakan numpy, pytoch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm
import numpy as np
```

```python
X = np.array([[92, 29, 10], [11, 40, 55], [17, 33, 22]])
X
```

```python
(92**2 + 29**2 + 10**2 + 11**2 + 40**2 + 55**2 + 17**2 + 33**2 + 22**2)**(1/2)
```

```python
np.linalg.norm(X)
```
