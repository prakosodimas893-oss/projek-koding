```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#1.buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array ([[1, 2, 3], [4, 5, 6]])
print(f"berikut matrix X dengan ukuran 2x3 menggunakan numpy {X}")
```

```python
#buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor ([[1, 2, 3], [4, 5, 6]])
print(f"berikut matrix X_pt dengan ukuran 2x3 menggunakan pytorch {X_pt}")
```

```python
#buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable ([[1, 2, 3], [4, 5, 6]])
print(f"berikut matrix X_tf dengan ukuran 2x3 menggunakan tensorflow{X_tf} ")
```

```python
#lakukan perkalian matrix X, X_pt dan X_tf dengan suatu skalar
#X
X*2
#X_pt
X_pt*2
#X_tf
X_tf*2
print(f"hasil perkalian X dengan skalar {X*2}")
print(f"hasil perkalian X_pt dengan skalar {X_pt*2}")
print(f"hasil perkalian X_tf dengan skalar {X_tf*2}")
```

```python
#lakukan penjumlahan matrix X, X_pt dan X_tf dengan suatu skalar
#X
X+2
#X_pt
X_pt+2
#X_tf
X_tf+2
print(f"hasil penjumlahan X dengan skalar {X+2}")
print(f"hasil penjumlahan X_pt dengan skalar {X_pt+2}")
print
print(f"hasil penjumlahan X_tf dengan skalar {X_tf+2}")
```

```python
#lakukan pengurangan matrix X, X_pt dan X_tf dengan suatu skalar
#X
A=X-2
#X_pt
B=X_pt-2
#X_tf-2
C=X_tf-2
print(f"hasil pengurangan X dengan skalar {A}")
print(f"hasil pengurangan X_pt dengan skalar {B}")
print(f"hasil pengurangan X_tf dengan skalar {C}")
```

```python
#2.buat matrix Y dengan ukuran 2x3 menggunakan numpy
Y = np.array ([[7, 8, 9], [10, 11, 12]])
print(f"berikut matrix Y dengan ukuran 2x3 menggunakan numpy {Y}")
```

```python
#buat matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow
Y_tf = tf.Variable ([[7, 8, 9], [10, 11, 12]])
print(f"berikut matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow {Y_tf}")
```
