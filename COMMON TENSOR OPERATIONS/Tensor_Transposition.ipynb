# Tensor Transposition

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
import numpy as np
import torch
import tensorflow as tf
```

```python
X = np.array([[29, 92, 85], [96, 93, 79]])
X
```

```python
X.T
```

# Using Torch

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[29, 92, 85], [96, 93, 79]])
X_pt
```

```python
X_pt.T
```

# Using Tensorflow

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[29, 92, 85], [96, 93, 79]])
X_tf
```

```python
tf.transpose(X_tf)
```
